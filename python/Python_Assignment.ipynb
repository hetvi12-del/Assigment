## 22) Reverse string if multiple of 4

```python

def rev_if_4(s):
    return s[::-1] if len(s) % 4 == 0 else s

print(rev_if_4("abcd"))
```

## 24) Insert string in middle

```python

def insert_mid(a, b):
    mid = len(a) // 2
    return a[:mid] + b + a[mid:]

print(insert_mid("<<>>", "Python"))
```

## 58) Combine values in list of dictionaries

```python

from collections import Counter
data = [{'item':'item1','amount':400},
        {'item':'item2','amount':300},
        {'item':'item1','amount':750}]
c = Counter()
for d in data:
    c[d['item']] += d['amount']
print(c)
```

## 61) Factorial function

```python

def factorial(n):
    return 1 if n == 0 else n * factorial(n-1)

print(factorial(5))
```

## 62) Number in range

```python

def in_range(n, start, end):
    return start <= n <= end

print(in_range(5,1,10))
```

## 63) Perfect number

```python

def perfect(n):
    return n == sum(i for i in range(1, n) if n % i == 0)

print(perfect(6))
```

## 64) Palindrome check

```python

def palindrome(s):
    return s == s[::-1]

print(palindrome("madam"))
```
